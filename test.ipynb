```python
import pandas as pd 
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/iris.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
feats = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
X= df[feats].values
y =df['Species'].values
```

```python
from sklearn.preprocessing import StandardScaler
# X_std = StandardScaler().fit_transform(X)
scaler =StandardScaler()
scaler.fit(X)
X_std =scaler.transform(X)
new_df = pd.DataFrame(data=X_std, columns= feats)
new_df.describe()
```

```python
#PCA
pca = PCA()
pca.fit(X)
Xpca=pca.transform(X)

pdf = pd.DataFrame(data=Xpca, columns=['pc1','pc2','pc3','pc4'])
pdf.describe()
```

```python
#Mengembalikan nilai ke Xtd
pca.inverse_transform(Xpca)
```

```python
#Dari scaler ke data asli
scaler.inverse_transform(pca.inverse_transform(Xpca))
```
